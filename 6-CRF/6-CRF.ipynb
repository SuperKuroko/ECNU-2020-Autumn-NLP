```python
result = open("result.txt","r", encoding="utf-8")
result=[e.strip() for e in result.readlines() if e.strip()]

count = 0  #记录正确个数
for e in result:
    words = e.split()         
    if words[-1] == words[-2]:  #比较预测和实际结果是否相同
        count += 1
print("Accuracy: %.4f" % (count/len(result)))


    
data = []
for line in result:
    words = line.split()
    if words[-1] != '0':
        data.append((words[0], words[-1]))

#把简写变成全程
def dict(s):
    if s == "PER":
        return "PERSON      : "
    elif s == "LOC":
        return "LOCATION    : "
    elif s == "ORG":
        return "ORGANIZATION: "
    return "MISC        : "

print("NER识别结果:",end = "")
Index = 0 #下标
for i in range(100):
    while data[Index][1][0] != "B":  #begin
        Index += 1
    length = 1
    while Index+length < len(data) and data[Index+length][1][0] == 'I':#end
        length += 1
    Type = dict(data[Index][1].split('-')[1])
    Word = [word[0] for word in data[Index:Index+length]]
    print("\n"+Type,end = " ")
    for w in Word:
        print(w,end = " ")
    Index += 1
```
